```python
from ultralytics import YOLO

# Load a model
# model = YOLO('yolov8n-pose.yaml')  # build a new model from YAML
model = YOLO('yolov8l-pose.pt')  # load a pretrained model (recommended for training)
# model = YOLO('yolov8n-pose.yaml').load('yolov8n-pose.pt')  # build from YAML and transfer weights

# Train the model
# results = model.train(data='coco8-pose.yaml', epochs=100, imgsz=640)
```

```python
# Validate the model
metrics = model.val()  # no arguments needed, dataset and settings remembered
metrics.box.map    # map50-95
metrics.box.map50  # map50
metrics.box.map75  # map75
metrics.box.maps   # a list contains map50-95 of each category
```

```python
# Predict with the model
model.predict('testJarak.mp4', save=True, imgsz=640, conf=0.25)  # predict on an image
```

```python
import cv2
from ultralytics import YOLO

model = YOLO('yolov8n.pt')

video_path = 0
cap = cv2.VideoCapture(video_path)
cv2.namedWindow('frame')

count = 1
while cap.isOpened():
  success, frame = cap.read()
  # sky = frame[350:450, 125:300]
  sky = frame
  if success:
    results = model(sky, save=True)
    annotated_frame = results[0].plot()
    print(annotated_frame)
    cv2.imshow("frame", annotated_frame)

    if cv2.waitKey(1) & 0xFF == ord("q"):
      break
  else:
    break

cap.release()
cv2.destroyAllWindows()
```
